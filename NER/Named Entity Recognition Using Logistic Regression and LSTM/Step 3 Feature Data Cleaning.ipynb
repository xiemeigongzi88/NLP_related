## clean test file

```python
df_1 = df_test.copy()

df_1 = df_1.drop(['hypernyms', 'hyponyms', 'holonyms', 'meronyms'], axis=1 )

df_1
```

```python
hypernyms_list =[]

for i, string in enumerate(df_test['hypernyms'].tolist()):
    #print(row)
    string = string[1:]
    #print(i, string)
    row_list = string.split()
    #print(row_list)
    
    empty=''
    
    for word in row_list:
        word=word[1:-2]+' '
        empty = empty + word 
        #print(string)
    
    #print(empty)
    #print(type(string), string)
    hypernyms_list.append(empty.strip())


hypernyms_list  # 46434
```

```python
df_1['hypernyms'] = hypernyms_list

df_1
```

```python
# 'hypernyms', 'hyponyms', 'holonyms', 'meronyms'

hyponyms_list =[]

for i, string in enumerate(df_test['hyponyms'].tolist()):
    #print(row)
    string = string[1:]
    #print(i, string)
    row_list = string.split()
    #print(row_list)
    
    empty=''
    
    for word in row_list:
        word=word[1:-2]+' '
        empty = empty + word 
        #print(string)
    
    #print(empty)
    #print(type(string), string)
    hyponyms_list.append(empty.strip())

print(len(hyponyms_list))
hyponyms_list  # 46434
```

```python
# hyponyms

df_1['hyponyms']= hyponyms_list

df_1
```

```python
# # 'hypernyms', 'hyponyms', 'holonyms', 'meronyms'

holonyms_list =[]

for i, string in enumerate(df_test['holonyms'].tolist()):
    #print(row)
    string = string[1:]
    #print(i, string)
    row_list = string.split()
    #print(row_list)
    
    empty=''
    
    for word in row_list:
        word=word[1:-2]+' '
        empty = empty + word 
        #print(string)
    
    #print(empty)
    #print(type(string), string)
    holonyms_list.append(empty.strip())

print(len(holonyms_list))
holonyms_list  # 46434
```

```python
df_1['holonyms']= holonyms_list

df_1
```

```python
# 'hypernyms', 'hyponyms', 'holonyms', 'meronyms'

meronyms_list =[]

for i, string in enumerate(df_test['meronyms'].tolist()):
    #print(row)
    string = string[1:]
    #print(i, string)
    row_list = string.split()
    #print(row_list)
    
    empty=''
    
    for word in row_list:
        word=word[1:-2]+' '
        empty = empty + word 
        #print(string)
    
    #print(empty)
    #print(type(string), string)
    meronyms_list.append(empty.strip())

print(len(meronyms_list))
meronyms_list  # 46434
```

```python
df_1['meronyms']= meronyms_list

df_1
```

```python
df_1.columns
```

```python
df_1 = df_1[['sentence','word','pos_tag', 'chunk_tag', 'lemmas',
       'hypernyms', 'hyponyms', 'holonyms', 'meronyms', 'ner_tag']]

df_1.reset_index(inplace=True)

df_1 = df_1.drop('index', axis=1)

df_1
```

```python
df_1.columns
```

```python
label =[]
for i in df_1['ner_tag'].tolist():
    try:
        if i=='O':
            label.append(0)
        elif i=='B-LOC':
            label.append(2)
        elif i=='I-LOC':
            label.append(2)
        elif i=='B-PER':
            label.append(1)
        elif i=='I-PER':
            label.append(1)
        elif i=='B-MISC':
            label.append(4)
        elif i=='I-MISC':
            label.append(4)
        elif i== 'B-ORG':
            label.append(3)
        elif i== 'I-ORG':
            label.append(3)
    except Exception as e:
        print(i+" "+str(e))
            
label.append(0)
```

```python
df_1['target']=label

df_1
```

```python
df_1.columns
```

```python
df_1 = df_1[['sentence', 'word','pos_tag', 'chunk_tag', 'lemmas', 'hypernyms', 'hyponyms', 'holonyms',
       'meronyms', 'target']]

df_1
```

```python
df_1.to_csv("test_extracted_features_step_3.csv", index=False)
```
